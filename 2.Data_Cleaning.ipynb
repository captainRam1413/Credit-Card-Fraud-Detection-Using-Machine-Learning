```python
import pandas as pd
```

```python
df=pd.read_csv("transactions_with_fraud.csv")
df
```

```python
df = df.drop(["transaction_id","name","email","phone"], axis=1)
```

```python
df
```

```python
print(df.select_dtypes(include=['object']).columns)
print(df.select_dtypes(include=['bool']).columns)
```

```python
from sklearn.preprocessing import LabelEncoder
encoders = {}

# Apply LabelEncoder to all string columns
for column in df.select_dtypes(include=['object']).columns:
    encoder = LabelEncoder()
    df[column] = encoder.fit_transform(df[column])
    encoders[column] = encoder
```

```python
for column in df.select_dtypes(include=['bool']).columns:
    encoder = LabelEncoder()
    df[column] = encoder.fit_transform(df[column])
    encoders[column] = encoder

# Show the transformed DataFrame
print(df)
```

```python
import joblib
joblib.dump(encoders, 'dynamic_encoders.pkl')
print("\nEncoders saved to 'dynamic_encoders.pkl'")
```

```python
nan_counts = df.isna().sum()

# Print the counts
print("Count of NaN values in each column:")
print(nan_counts)
```
